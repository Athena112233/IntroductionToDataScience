```python
#merge, join, concat https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html
happy2015 = pd.read_csv('data/happiness2015.csv') #happiness score
happy2016 = pd.read_csv('data/happiness2016.csv') #happiness score
happy2017 = pd.read_csv('data/happiness2017.csv') #happiness score
happy2018 = pd.read_csv('data/happiness2018.csv') #score
happy2019 = pd.read_csv('data/happiness2019.csv') #score
happy2020 = pd.read_csv('data/happiness2020.csv') #ladder score
```

```python
happy2015 = happy2015.rename(columns  = {'Happiness Score': 'Score', 'Happiness Rank': 'Rank', })[['Country', 'Region', 'Score']]
happy2016 = happy2016.rename(columns  = {'Happiness Score': 'Score'})[['Country', 'Region', 'Score']]
happy2017 = happy2017.rename(columns  = {'Happiness.Score': 'Score'})[['Country', 'Score']]
happy2018 = happy2018.rename(columns  = {'Country or region': 'Country'})[['Country', 'Score']]
happy2019 = happy2019.rename(columns  = {'Country or region': 'Country'})[['Country', 'Score']]
happy2020 = happy2020.rename(columns = {'Ladder score': 'Score', 'Country name': 'Country', 'Regional indicator': 'Region'})[['Country', 'Region', 'Score']]
```

```python
happy2015.shape #158 countries
happy2016.shape #157 countries
happy2017.shape #155 countries
happy2018.shape #156 countries
happy2019.shape #156 countries
happy2020.shape #153 countries
```

```python
lst2020 = {'Taiwan Province of China': 'Taiwan', 'Hong Kong S.A.R. of China': 'Hong Kong', 'South Sudan': 'Sudan'}
```

```python
#column cleaning
happy2020['Country'] = happy2020['Country'].replace(lst2020)
```

```python
temp = pd.merge(happy2015, happy2020, right_on = 'Country', left_on = 'Country', suffixes=('2015', '2020')) #previous merge 147
temp
```

```python
top20 = temp.iloc[:20]
top20
```

```python
sns.set(style="whitegrid")
f, ax = plt.subplots(figsize=(6, 10))
sns.set_color_codes("pastel")
sns.barplot(x = 'Score2015', y = 'Country', data = top20)
```
