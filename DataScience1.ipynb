```python
#import libraries
import pandas as pd #data analysis, organizational, manipulation library
import numpy as np #math library
import matplotlib.pyplot as plt #plotting library - data visualization
import seaborn as sns #plotting library - data visalizations
```

```python
#import data files
happy2020 = pd.read_csv('data/happiness2020.csv')
```

```python
#step1: examine the data file
#things to look for: size, column names, data type
happy2020
```

```python
#how to dispaly partial of the dataframe
happy2020.head(5)
```

```python
#how many rows x columns I have aka dimension
#return a tuple
happy2020.shape
```

```python
#how to get specific columns
#only look at country names
happy2020['Country name'] #series
```

```python
happy2020[['Country name']] #dataframe
```

```python
#how to get specific value
#index 3, iceland, get regional indicator
#iloc --> location by index
happy2020.iloc[3]['Regional indicator'] #on dataframe
```

```python
happy2020['Regional indicator'].iloc[3] #on series
```

```python
#how to get rows by specific values
#loc --> find location
#pandas use True/False selector --> if row is classified as false, it's gonna get filtered out
#if row is classified as true, keep that row

happy2020.loc[happy2020['Country name'] == 'China']
```

```python
#find all the countries that's in North America
happy2020.loc[happy2020['Regional indicator'] == 'North America and ANZ']
```

```python
#locate with multiple statements
happy2020.loc[(happy2020['Country name'] == 'China') | (happy2020['Country name'] == 'United States')]
```

```python
#numeric
happy2020.loc[happy2020['Ladder score'] >= 7]
```
